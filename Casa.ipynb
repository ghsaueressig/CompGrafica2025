```python
import pygame
import sys
import matplotlib.pyplot as plt

# Inicializando o pygame
pygame.init()

# Definindo as dimensões da janela
width, height = 800, 600
screen = pygame.Surface((width, height))  # Usando uma superfície em vez de uma janela

# Definindo as cores
PRETO = (0, 0, 0)
VERDE = (0, 255, 0)
MARROM = (139, 69, 19)
VERMELHO = (255, 0, 0)
AMARELO = (255, 255, 0)
CEU = (135, 206, 235)  # Fundo azul (céu)

# Preenchendo o fundo com a cor do céu
screen.fill(CEU)

# Desenha o chão verde
pygame.draw.rect(screen, VERDE, (0, 400, 800, 200))

# Desenha a base da casa (retângulo marrom)
pygame.draw.rect(screen, MARROM, (300, 300, 200, 150))

# Desenha o telhado (triângulo vermelho)
pygame.draw.polygon(screen, VERMELHO, [(300, 300), (400, 200), (500, 300)])

# Desenha a porta (retângulo preto)
pygame.draw.rect(screen, PRETO, (370, 350, 60, 100))

# Desenha a janela (retângulo amarelo)
pygame.draw.rect(screen, AMARELO, (320, 320, 40, 40))

# Linhas da janela
pygame.draw.line(screen, PRETO, (340, 320), (340, 360), 2)  # Linha vertical
pygame.draw.line(screen, PRETO, (320, 340), (360, 340), 2)  # Linha horizontal

# Desenha a chaminé
pygame.draw.rect(screen, PRETO, (430, 220, 20, 60))

# Salva a tela como uma imagem
pygame.image.save(screen, "/content/casa_com_telhado.png")

# Converte a imagem salva para exibição no Colab
img = plt.imread('/content/casa_com_telhado.png')

# Exibe a imagem usando Matplotlib
plt.imshow(img)
plt.axis('off')  # Desativa os eixos
plt.show()

# Finaliza o pygame
pygame.quit()
```
